B. Import Datasets

```python
# Load MTN Telecom dataset
df1 = pd.read_csv('Telcom_dataset.csv')
df1.head()
```

```python
# Load MTN Telecom dataset 2
df2 = pd.read_csv('Telcom_dataset2.csv')
df2
```

```python
# Load MTN Telecom dataset 3
df3 = pd.read_csv('Telcom_dataset3.csv')
df3.head()
```

```python
# Load MTN Location dataset
df4 = pd.read_excel('cells_geo.xlsx')
df4.head()
```

DATA PREPARATION

1.1 Telecom Dataset

```python
# We will start by renaming telecom dataset columns to allow for proper merging of files
df1.rename(columns={'PRODUTC':'PRODUCT','DATETIME': 'DATE_TIME'},inplace = True)
df2.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT'},inplace = True)
df3.rename(columns={"CELLID" : "CELL_ID", 'SIET_ID' : 'SITE_ID'},inplace = True)
```

```python
# We are going to merge the three telecom datasets(df,df2 and df3) into one :
Telecom1=pd.concat((df1, df2, df3))
Telecom1
```

```python
# Drop columns
Telecom1.drop(columns=['COUNTRY_A','COUNTRY_B','DW_A_NUMBER_INT','DW_B_NUMBER_INT'],inplace=True ,axis=1 )
```

```python
# Check id columns have been dropped
Telecom1
```

```python
# Delete rows containing Null Values
Telecom10=Telecom1[Telecom1.VALUE != 0]
Telecom10
```

```python
# Delete rows containing duplicate values
Telecom11=Telecom10.drop_duplicates(['CELL_ID'])
Telecom11
```

```python
# Remove / and + from the Cell_ID Column
Telecom11['CELL_ID']=Telecom11['CELL_ID'].str.replace(r'\W','')
Telecom11
```

1.2  Cells_Geo Dataset

```python
# Drop columns
df4.drop(columns=['Unnamed: 0','LOCALISATION','DECOUPZONE','ZONENAME','LATITUDE','LONGITUDE'],inplace=True ,axis=1 )
df4
```

```python
# Delete all rows with Nullvalues

df5=df4.dropna(how='any')
df5
```

```python
df5.rename(columns={"SITE_CODE" : "SITE_ID"} ,inplace = True)
```

```python
# Merge Telcom Dataset with Cells_Geo Dataset

df_merge = pd.merge(Telecom11,df5 ,how='inner',on='SITE_ID')
df_merge
```
